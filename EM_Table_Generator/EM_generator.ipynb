```python
import numpy as np
import sys
!{sys.executable} -m pip install scipy

from scipy.stats import binom

# Data from the video/image
heads = np.array([3, 4, 7, 9])
tails = np.array([7, 6, 3, 1])
n_flips = 12

# Initial parameters (as in video)
theta_A = 0.7
theta_B = 0.2

print("=== EM Algorithm Step by Step (Exact Video Style) ===\n")
print(f"Initial: θ_A = {theta_A:.2f}    θ_B = {theta_B:.2f}\n")

max_iter = 10
tol = 1e-5

for it in range(max_iter):
    print(f"{'='*90}")
    print(f"Iteration {it+1}")
    print(f"{'='*90}")
    
    # E-Step
    lik_A = binom.pmf(heads, n_flips, theta_A)
    lik_B = binom.pmf(heads, n_flips, theta_B)
    total_lik = lik_A + lik_B
    
    prob_A = lik_A / total_lik
    prob_B = lik_B / total_lik
    
    # Attributed Heads/Tails
    attr_H_A = prob_A * heads
    attr_T_A = prob_A * tails
    attr_H_B = prob_B * heads
    attr_T_B = prob_B * tails
    
    # Print Table Header
    print(f"{'Heads':<6} {'Tails':<6} {'Lik. Coin A':<15} {'Lik. Coin B':<15} "
          f"{'Prob A':<10} {'Prob B':<10} {'Attr to A':<20} {'Attr to B':<20}")
    print("-" * 110)
    
    # Print each sequence row
    for i in range(4):
        print(f"{heads[i]:<6} {tails[i]:<6} {lik_A[i]:<15.8f} {lik_B[i]:<15.8f} "
              f"{prob_A[i]:<10.4f} {prob_B[i]:<10.4f} "
              f"{attr_H_A[i]:.3f}H {attr_T_A[i]:.3f}T    "
              f"{attr_H_B[i]:.3f}H {attr_T_B[i]:.3f}T")
    
    # M-Step
    total_H_A = np.sum(attr_H_A)
    total_T_A = np.sum(attr_T_A)
    total_H_B = np.sum(attr_H_B)
    total_T_B = np.sum(attr_T_B)
    
    new_theta_A = total_H_A / (total_H_A + total_T_A)
    new_theta_B = total_H_B / (total_H_B + total_T_B)
    
    print("\n" + "-"*60)
    print(f"Total Attributed to Coin A: {total_H_A:.3f}H   {total_T_A:.3f}T")
    print(f"Total Attributed to Coin B: {total_H_B:.3f}H   {total_T_B:.3f}T")
    print(f"\nNew θ_A = {total_H_A:.3f} / ({total_H_A:.3f} + {total_T_A:.3f}) = {new_theta_A:.4f}")
    print(f"New θ_B = {total_H_B:.3f} / ({total_H_B:.3f} + {total_T_B:.3f}) = {new_theta_B:.4f}")
    
    if abs(theta_A - new_theta_A) < tol and abs(theta_B - new_theta_B) < tol:
        print(f"\n✅ CONVERGED at Iteration {it+1}!")
        break
    
    theta_A = new_theta_A
    theta_B = new_theta_B

print("\n" + "="*60)
print("FINAL RESULT:")
print(f"θ_A (Coin A) = {theta_A:.4f}")
print(f"θ_B (Coin B) = {theta_B:.4f}")
```
